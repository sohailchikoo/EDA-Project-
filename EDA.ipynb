**Descriptive Statistics: Calculate basic statistics (mean, median, mode, standard deviation).**

**Mean**-In above data we should calculate mean for columns containing numbers like Qauntity, Price per unit,Total Amount

```python
mean_values = df[['Quantity', 'Price per Unit', 'Total Amount']].mean()
mean_values
```

**Median**-In above data we should calculate median for columns containing numbers like Qauntity, Price per unit,Total Amount .

```python
median_values = df[['Quantity', 'Price per Unit', 'Total Amount']].median()
median_values
```

**Mode**-In above data we should calculate mode for columns containing numbers like Qauntity, Price per unit,Total Amount .

```python
mode_values = df[['Quantity', 'Price per Unit', 'Total Amount']].mode().iloc[0]
mode_values
```

**Standard Deviation**-In above data we should calculate std

**Time Series Analysis: Analyze sales trends over time using time series techniques.**


```python
 #Convert the Date Column to a Datetime Object

 df['Date'] = pd.to_datetime(df['Date'])
 df['Date']
```

```python
# Resample the data to get monthly sales
monthly_sales = df['Total Amount'].resample('M').sum()
monthly_sales
```

```python
import matplotlib.pyplot as plt

# Plot the monthly sales trends
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

**Customer and Product Analysis: Analyze customer demographics and purchasing behavior.**

** Customer Demographics Analysis**
Summary Statistics: Calculate basic statistics (mean, median, mode, standard deviation) for demographic variables like age and income.

Segmentation: Group customers by demographics (e.g., age groups, gender, income brackets) to identify patterns and trends.

```python
# Example: Segmenting customers by age group
df['Age Group'] = pd.cut(df['Age'], bins=[18, 30, 40, 50, 60, 100], labels=['18-29', '30-39', '40-49', '50-59', '60+'])
age_group_summary = df.groupby('Age Group').size()
```

```python
# Example: Age group distribution
import matplotlib.pyplot as plt
age_group_summary.plot(kind='bar')
plt.title('Customer Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Example: Product sales analysis
product_sales = df.groupby('Customer ID')['Total Amount'].sum().sort_values(ascending=False)
product_sales.head()
```

```python
import matplotlib.pyplot as plt

# Example data
age_group_summary = pd.DataFrame({
    'Age Group': ['18-29', '30-39', '40-49', '50-59', '60+'],
    'Number of Purchases': [150, 200, 180, 160, 130]
})

plt.figure(figsize=(10, 6))
plt.bar(age_group_summary['Age Group'], age_group_summary['Number of Purchases'], color='skyblue')
plt.title('Number of Purchases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
monthly_sales = pd.Series([1200, 1500, 1700, 1600, 1800, 2000, 2100, 1900, 2200, 2300, 2400, 2500],
                          index=pd.date_range(start='2023-01-01', periods=12, freq='M'))

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
purchase_matrix = pd.DataFrame({
    'Product 1': [10, 15, 5, 8],
    'Product 2': [20, 18, 7, 12],
    'Product 3': [30, 22, 10, 15]
}, index=['Age 18-29', 'Age 30-39', 'Age 40-49', 'Age 50+'])

plt.figure(figsize=(10, 6))
sns.heatmap(purchase_matrix, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Purchase Frequency by Age Group and Product')
plt.xlabel('Product')
plt.ylabel('Age Group')
plt.show()
```
